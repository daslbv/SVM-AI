# Text PreProcessing
Bagian Kolom 'teks'

```python
import nltk.corpus
from nltk.corpus import stopwords
```

Bagian text Proses mengubah seluruh huruf menjadi huruf kecil, menghapus elemen yang kosong, huruf yang double, dan angka

```python
import re
def  clean_text(data, text_field, new_text_field_name):
    data[new_text_field_name] = data[text_field].str.lower()
    data[new_text_field_name] = data[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))  # remove numbers
    data[new_text_field_name] = data[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem))
    data[new_text_field_name] = data[new_text_field_name].apply(lambda elem: re.sub(r"(.)\1{2,}", r"\1\1", elem))  # Menghapus huruf yang triple
    data[new_text_field_name] = data[new_text_field_name].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3])) #menghapus kata pendek
    return data
```

Text-Prepocessing membuat kolom baru untuk menyimpan data yang sudah dibersihkan

```python
data.dropna(subset=['text'], inplace=True)
data['text_clean'] = data['text'].str.lower()
data['text_clean']
data_clean = clean_text(data, 'text', 'text_clean')
data_clean.head()
```

Menghilangkan kata yang tidak bermakna menggunakan library stopwords

```python
stop = stopwords.words('english')
data_clean['text_StopWord'] = data_clean['text_clean'].apply(lambda x:' '.join([word for word in x.split() if word not in (stop)]))
data_clean.head(5)
```

Bagian kolom 'description'

```python
data['description_clean'] = data['description'].str.lower()
data['description_clean']
data_description = clean_text(data, 'description', 'description_clean')
data_description.head(5)
```

```python
#proses stop word atau menghapus kata-kata yang tidak perlu
stop = stopwords.words('english')
data_description['description_StopWord'] = data_description['description_clean'].apply(lambda x:' '.join([word for word in x.split() if word not in (stop)]))
data_description.head(5)
```
